Load Amazon Data into Spark DataFrame



```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Office_Products_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get("amazon_reviews_us_Office_Products_v1_00.tsv.gz"), sep="\t", header=True, inferSchema=True)
df.show()
```

### Create DataFrames to match tables

```python
from pyspark.sql.functions import to_date
# Read in the Review dataset as a DataFrame. DONE IN CODE BLOCK ABOVE.
```

```python
# Recreate the vine_table DataFrame
vine_df = df.select(["review_id", "star_rating", "helpful_votes", "total_votes", "vine", "verified_purchase"])
vine_df.show()
```

##### Deliv 2 Q1: Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20 to pick reviews that are more likely to be helpful and to avoid having division by zero errors later on.

```python
# Deliv 2 Q1: Filter data & create new DataFrame to show only reviews with 20 or more total votes
total_vote_20_df = vine_df.filter("total_votes >= 20")
total_vote_20_df.show()
```

```python
# Deliv 2 Q2: Filter df created in step 1 & create new DataFrame to retrive rows where helpful_votes/total_votes > = 50%
vote_breakdown_df = total_vote_20_df.filter("(helpful_votes/total_votes) >= 0.5")
vote_breakdown_df.show()
```

```python
# Deliv 2 Q3: Filter df created in step 2 & create new DataFrame to retrive rows where review was written as part of Vine program (paid)
paid_reviews_df = vote_breakdown_df.filter("vine == 'Y'")
paid_reviews_df.show()
```

```python
# Deliv 2 Q4: Repeat step 3: Filter df created in step 2 & create new DataFrame to retrive rows where review was not part of Vine program (unpaid)
unpaid_reviews_df = vote_breakdown_df.filter("vine == 'N'")
unpaid_reviews_df.show()
```

##### Deliv 2 Q5: Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

```python
# Total number of reviews
total_reviews = vote_breakdown_df.count()
total_reviews
```

```python
# Total number of 5-Star reviews
five_star_reviews = vote_breakdown_df.filter("star_rating == 5")
five_star_count = five_star_reviews.count()

five_star_count
```

```python
# Number of Vine (paid) reviews -- can use the df already created in step 3 above
paid_reviews_count = paid_reviews_df.count()
paid_reviews_count
```

```python
# Of the Vine (paid) reviews, how many are 5 stars
paid_five_star = paid_reviews_df.filter("star_rating == 5")
paid_five_star_count = paid_five_star.count()

paid_five_star_count
```

```python
# Percentage of Vine reviews that were 5 stars
paid_fivestar_percent = (paid_five_star_count / paid_reviews_count) * 100
paid_fivestar_percent
```

```python
# Number of non-Vine (unpaid) reviews
unpaid_reviews_count = unpaid_reviews_df.count()
unpaid_reviews_count
```

```python
# Of the non-Vine (unpaid) reviews, how many are 5 stars
unpaid_five_star = unpaid_reviews_df.filter("star_rating == 5")
unpaid_five_star_count = unpaid_five_star.count()

unpaid_five_star_count
```

```python
# Percentage of non-Vine reviews that were 5 stars
unpaid_fivestar_percent = (unpaid_five_star_count / unpaid_reviews_count) * 100
unpaid_fivestar_percent
```
